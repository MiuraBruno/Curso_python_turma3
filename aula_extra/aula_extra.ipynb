```python
#operador ternário
```

```python
def par_impar(n):
    if n % 2 == 0:
        return 'Par'
    else:
        return 'Impar'
```

```python
par_impar_2 = lambda n: "Par" if n % 2 == 0 else "Impar"
```

```python
par_impar(3)
```

```python
par_impar_2(3)
```
