```python
leds = {'1':2
        ,'2':5
        ,'3':5
        ,'4':4
        ,'5':5
        ,'6':6
        ,'7':3
        ,'8':7
        ,'9':6
        ,'0':6
       }
```

```python
qtd = int(input())
```

```python
for i in range(qtd):
    numero = input()
    total = 0
    for j in numero:
        total = total + leds[j]
    print('%d leds' %(total))
```
